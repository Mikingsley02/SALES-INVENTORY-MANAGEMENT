# Project Title: SALES INVENTORY MANAGEMENT

```python
from datetime import date

# List of categories

categories = ["Beverages","PhoneAccessories","Toiletries","Pastry","Cosmetics"]

# Categories and their products

category_items = {"Beverages":["chocolateDrinks", "coffee", "tea", "soyDrinks", "pop","soda"],
    "PhoneAccessories": ["carryingCase", "earpieces", "screenGuards"],
    "Toiletries": ["toiletPaper", "bodySoap", "scrubs", "bodyCreme", "shampoo"],
    "Pastry": ["pizza", "burgers", "donuts", "muffins", "cheesecakes"],
    "Cosmetics": ["perfumes", "vanishes", "nailPolish", "deodorants", "facialScrubs"]}


# All Category Inventories using dictionaries

inventories = {
    "beverages_inventory":{
        "chocolateDrinks": {
            "purchasePrice":10,
            "price":20,
            "quantity":20
        },
        "coffee": {
            "purchasePrice":5,
            "price":15,
            "quantity":30
        },
        "tea":{
            "purchasePrice":5,
            "price":10,
            "quantity":20
        },
        "soyDrinks":{
            "purchasePrice":15,
            "price":25,
            "quantity":40
        },
        "pop":{
            "purchasePrice":20,
            "price":30,
            "quantity":10
        },
        "soda":{
            "purchasePrice":2,
            "price":6,
            "quantity":50
        }
    },
    "phoneaccessories_inventory":{
        "carryingCase": {
            "purchasePrice":50,
            "price":70,
            "quantity":7
        },
        "earpieces": {
            "purchasePrice":80,
            "price":100,
            "quantity":25
        },
        "screenGuards":{
            "purchasePrice":100,
            "price":120,
            "quantity":35
        }
    },
    "toiletries_inventory":{
        "toiletPaper": {
            "purchasePrice":50,
            "price":70,
            "quantity":20
        },
        "bodySoap": {
            "purchasePrice":20,
            "price":50,
            "quantity":40
        },
        "scrubs":{
            "purchasePrice":30,
            "price":60,
            "quantity":35
        },
        "bodyCreme":{
            "purchasePrice":20,
            "price":60,
            "quantity":45
        },
        "shampoo":{
            "purchasePrice":60,
            "price":100,
            "quantity":10
        }
    },
    "pastry_inventory":{
        "pizza": {
            "purchasePrice":10,
            "price":20,
            "quantity":25
        },
        "burgers": {
            "purchasePrice":20,
            "price":50,
            "quantity":50
        },
        "donuts":{
            "purchasePrice":5,
            "price":10,
            "quantity":50
        },
        "muffins":{
            "purchasePrice":5,
            "price":10,
            "quantity":50
        },
        "cheesecakes":{
            "purchasePrice":10,
            "price":20,
            "quantity":20
        }
    },
    "cosmetics_inventory":{
        "perfumes": {
            "purchasePrice":120,
            "price":150,
            "quantity":15
        },
        "vanishes": {
            "purchasePrice":80,
            "price":100,
            "quantity":10
        },
        "nailPolish":{
            "purchasePrice":130,
            "price":150,
            "quantity":25
        },
        "deodorants":{
            "purchasePrice":100,
            "price":130,
            "quantity":35
        },
        "facialScrubs":{
            "purchasePrice":60,
            "price":80,
            "quantity":24
        }
    }
}

# function to format the product names being provided in
# a manner that relates to the inventories

def productNameFormatter(productName: str):
    #if the product name is a two letter word format it will be splitted and then combined
    if(" " in productName):
        productName = productName.split(" ")
        productName = productName[0].lower() + productName[1].title()
        return productName
    else:
        productName = productName.lower()
        return productName

# function to check if a product exists in the inventories
def prodExist(prod):
    for key, value in category_items.items():

        if prod in value:
            if key == 'Beverages':
                return [key, inventories["beverages_inventory"][prod]]
            elif key == 'PhoneAccessories':
                return [key, inventories["phoneaccessories_inventory"][prod]]
            elif key == 'Toiletries':
                return [key, inventories["toiletries_inventory"][prod]]
            elif key == 'Pastry':
                return [key, inventories["pastry_inventory"][prod]]
            elif key == 'Cosmetics':
                return [key, inventories["cosmetics_inventory"][prod]]
        
    return False

#function to check if beverages exist
bevexist = input('Want to check if a product exist? please enter product name: ')
result = prodExist(bevexist)
print("product details")
print(result)

# function to generate receipt for purchase
def receiptGenerator(receipt):
    total = 0
    if len(receipt) > 0:
        print("RECEIPT")
        print("---------")
        for i in receipt:
            print(f"Product: {i['productName']} | Quantity: {i['quantity']} | SubTotal: {i['total']}")
            total = total + i['total']
        print(f"Total: {total}")
    else:
        print("YOUR RECEIPT IS EMPTY")


# function to track and update the sales records

def addToSalesRecord(salesRecords,quantity, result, sales_price, productName,total_cost):
    sales_details = {}

    sales_details["quantity"] = quantity
    sales_details["productPrice"] = result[1]["purchasePrice"]
    sales_details["soldFor"] = sales_price
    sales_details["category"] = result[0]
    sales_details["productName"] = productName
    sales_details["date"] = f"{date.today()}"
    sales_details["totalCost"] = total_cost

    id = str(len(salesRecords) + 1)

    salesRecords[id] = sales_details
    
    return salesRecords

    

# To fulfill the condition specified question number six(6)
#b. Show total sales by product

#c. Show total sales by category

#d. Show total sales for each day

#function that answers the option b,c,d of requirement 6 as stated in the inventory requirements
def onEachSales(salesRecords):

    #To show total sales for each categories
    print("TOTAL SALES BY CATEGORY")
    print("_______________________")
    for category in categories:
        catTotal = 0
        for record in salesRecords.values():
            if category == record["category"]:
                catTotal += record["totalCost"]
        print(f"{category} ==> {catTotal}")

    #To show total sales for each product
    print("\n")
    print("TOTAL SALES BY PRODUCT")
    print("_______________________")
    
    newarray = []
    for i in category_items.values():
        newarray += i
    for products in newarray:
        productTotal = 0
        for record in salesRecords.values():
            if products == record["productName"]:
                productTotal += record["totalCost"]
        print(f"{products} ==> {productTotal}")

    #To show total sales for each date recorded in the sales record dictionary
    print("\n")
    print("TOTAL SALES BY DATE")
    print("_______________________")
    #I used sets to remove repetetions
    dateArray = set()
    for i in salesRecords.values():
        dateArray.add(i["date"])

    for date in dateArray:
        productTotal = 0
        for record in salesRecords.values():
            if date == record["date"]:
                productTotal += record["totalCost"]
        print(f"{date} ==> {productTotal}")

### This is the sales function that controls the sales of products

def productPurchase():
    #receipt tracker
    receipt = []
    #sale record book or sales records tracker
    salesRecords = {}

    #Added the while loop so has to perform multiple sales
    while(True):
        #Accept inputs from the user
        productName = input("Enter the name of the product: ")

        #process the input with the productNameFormatter()
        productName = productNameFormatter(productName)

        #check if the product exist in the inventories
        result = prodExist(productName)

        #if it does
        if result:
            #display the product details(category name and other product details)
            print(f"Product details: {result}")

            #get the unit price of the product from inventories
            sales_price = int(result[1]["price"])
            print(f"{productName} is sold at {sales_price} per unit")

            #accept quantity from the user
            quantity = int(input("Enter the number of items: "))

            #ensure that the quantity available is not more than the user inputted qunatity 
            if quantity >= int(result[1]["quantity"]):
                print(f"PLEASE ENTER AN ITEM NUMBER LESS THAN {result[1]['quantity']}")
                print("\n")
                #if it is more than the available qunatity generate a receipt and end the sale process
                receiptGenerator(receipt)
                break

            #function to check if the remaining product quantity is less than 5
            if int(result[1]["quantity"])-quantity < 5:
                print(f"[ALERT] {productName} NOW AS LESS THAN 5 AVAILABLE UNITS")

            #calculate the total cost of everything(unitPrice * quantity)
            total_cost = sales_price * quantity
            print(f"Total cost of purchase: {total_cost}")
            # print("\n")

            #confirm the sale
            confirm_Sell = input("PLEASE CONFIRM ORDER (Y|y/N|n): ")
            print("\n")

            if confirm_Sell in ["Y","y"]:
                #updating the sales records tracker
                salesRecords = addToSalesRecord(salesRecords,quantity, result, sales_price, productName,total_cost)
                #if confirmed
                receipt.append({"productName":productName,"quantity":quantity,"total":total_cost})
            else:
                #if not comfirmed
                print("ORDER CANCELLED")
                break
        else:
            #if product was not found
            print(f"product {productName} doesn't exist")

        #do you want to continue shopping ???
        continueShopping = input("Do you want to continue shopping?(Y|y/N|n) : ")
        #yes i want to continue shopping
        if continueShopping in ["Y","y"]:
            continue
        #no i dont want to continue shopping
        else:
            print("SHOPPING ENDED(CHECKED OUT)")
            print("\n")
            #display receipt
            receiptGenerator(receipt)
            print('\n')
            #calls the total sales by category, product or date function
            onEachSales(salesRecords)
            break

productPurchase()

#function to restock the inventories

def restock(prod, qty):
    prod = productNameFormatter(prod)

    #To convert user input to integer
    qty = int(qty)

    data = inventories

    for key, value in category_items.items():
        if prod in value:
            if key == 'Beverages':
                oldQty =  int(data["beverages_inventory"][prod]["quantity"])
                data["beverages_inventory"][prod]["quantity"] = int(data["beverages_inventory"][prod]["quantity"]) + qty
                
                print(f"quantity of {prod} increase from {oldQty} to {data['beverages_inventory'][prod]['quantity']}")
                
            elif key == 'PhoneAccessories':
                oldQty =  int(data["phoneaccessories_inventory"][prod]["quantity"])

                data["phoneaccessories_inventory"][prod]["quantity"] = int(data["phoneaccessories_inventory"][prod]["quantity"]) + qty
                
                print(f"quantity of {prod} increase from {oldQty} to {data['phoneaccessories_inventory'][prod]['quantity']}")

            elif key == 'Toiletries':
                oldQty =  int(data["toiletries_inventory"][prod]["quantity"])

                data["toiletries_inventory"][prod]["quantity"] = int(data["toiletries_inventory"][prod]["quantity"]) + qty
                
                print(f"quantity of {prod} increase from {oldQty} to {data['toiletries_inventory'][prod]['quantity']}")

            elif key == 'Pastry':
                oldQty =  int(data["pastry_inventory"][prod]["quantity"])

                data["pastry_inventory"][prod]["quantity"] = int(data["pastry_inventory"][prod]["quantity"]) + qty
                
                print(f"quantity of {prod} increase from {oldQty} to {data['pastry_inventory'][prod]['quantity']}")

            elif key == 'Cosmetics':               
                oldQty =  int(data["cosmetics_inventory"][prod]["quantity"])
                
                data["cosmetics_inventory"][prod]["quantity"] = int(data["cosmetics_inventory"][prod]["quantity"]) + qty
                
                print(f"quantity of {prod} increase from {oldQty} to {data['cosmetics_inventory'][prod]['quantity']}")

                

toRestock = input("DO YOU WANT TO RESTOCK INVENTORIES:(Y|y|N|n) ")

if toRestock in ["Y","y"]:
    productName = input("Please enter the name of product to restock: ")
    quantity = input("Enter the number of products to add: ")

    restock(productName,quantity)
else:
    print("No Restocking")

```
